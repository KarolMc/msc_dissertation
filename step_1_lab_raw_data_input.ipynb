```python
import os

def folder_contents_csv(category, path):
  file_paths = []
  table_names = []
  for file in os.listdir(path):
      if file.endswith(".csv"):
        file_paths.append(os.path.join(path, category, file))

  for path in file_paths: 
    path = path.split('.', 1)[0]
    name = path.split('/')[-1]
    table_names.append(name)

  return file_paths, table_names
```

```python
def diabetes_check(row):
  if row['LBXGH'] >= 6.5 or row['LBXGLU'] >= 126 or row['LBXGLT'] >= 200:
    return 'Diabetic on one or more tests'
  elif (row['LBXGH'] >= 5.7 and row['LBXGH'] <6.5) or (row['LBXGLU'] >= 100 and row['LBXGLU'] < 126) or (row['LBXGLT'] >= 140 and row['LBXGLT'] < 200):
    return 'Prediabetes on one or more tests' 
  else:
    return 'Normal'
```
